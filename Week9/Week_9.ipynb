## 03 Solution part 1

1. Create a python notebook, and download the dataset from the link above.

```python
import urllib.request

print('Beginning file download with urllib...')

url = "https://snap.stanford.edu/data/wiki-Vote.txt.gz"
urllib.request.urlretrieve(url, "./downloaded_files/Wiki-Vote.txt.gz")

print("Download with urllib finished...")
```

2. Use python to create a new file and copy all the data from the .txt.gz into a .txt file.

```python
import gzip
import shutil

# open the unzipped file with flie handler inp_f
with gzip.open('./downloaded_files/Wiki-Vote.txt.gz', 'rb') as in_f:
     with open('./downloaded_files/Wiki-Vote.txt',"wb") as out_f:
         # use shutil to copy the file objec
         shutil.copyfileobj(in_f,out_f)
```

3. Use networkx to add the nodes from the file we just created: https://networkx.org/documentation/stable/reference/classes/graph.html

```python
import networkx as nx

g = nx.read_edgelist('./downloaded_files/Wiki-Vote.txt')
print(nx.info(g))
```

4. Make a random number generator that is able to create a number between 1 and the length of your Undirected graphs.

```python
import random

def random_number_generator(graph):
    return random.choice(list(g.nodes()))

print(random_number_generator(g))
```

5. Print all the neighbouring nodes, of the randomly generated id.

```python
for neighbour in g.neighbors(random_number_generator(g)):
    print(neighbour)
```

6. Make a visual presentation of the randomly generated id in the notebooks. nx.draw()

```python
nx.draw_circular(g.subgraph(g[random_number_generator(g)]), with_labels=True)
```
